## Filtros

### Filtros con condiciones lógicas

```python
# Filtrar los registros que sean Movie
#df.query('''
 # type == "Movie"
#''')
df[df["type"] =="Movie"]
```

```python
df.head(2)
```

```python
# Los registros cuyo campo country es Argentina
#df.query('country == "Argentina"')
df[df["country"]=="Argentina"]
```

```python
# Los registros (movies/series) que cuyo año de "release" es anterior al 2020
df.query('type in ["Movie",  "series" ] and release_year < 2020 ').sort_values("release_year", ascending = False)

```

```python
# Verificar el dtype de la columna "release_year"
df.release_year.dtype
df.date_added.dtype
```

```python
# Los registros (movies/series) que cuyo date_added es anterior al 2020
df.loc[df["date_added"].dt.year < 2020, "date_added":].sort_values("date_added", ascending= False)
```
